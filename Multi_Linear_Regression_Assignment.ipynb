```python
import pandas as pd
Startups = pd.read_csv("Startups.csv")
Startups = Startups.rename(columns={"R&D Spend": "Research"})
Startups = Startups.rename(columns={"Marketing Spend": "Marketing"})
```

```python
Startups.info()
```

```python
Startups.isnull().sum()
```

```python
Startups.corr
```

```python
sns.set_style(style='darkgrid')
sns.pairplot(Startups)
```

```python
import statsmodels.api as sm
import pandas as pd
```

```python
X = Startups[['Research', 'Administration', 'Marketing']] 
y = Startups['Profit'] 
model = sm.OLS(y, sm.add_constant(X)).fit()
```

```python
fig, ax = plt.subplots(figsize=(12,8))
fig = sm.graphics.influence_plot(model, ax=ax, criterion="cooks")
```

```python
model = smf.ols('Profit ~ Administration + Research + Marketing', data=Startups).fit()
```

```python
model.fittedvalues
```

```python
model.params
```

```python
Startups
```

```python
Startups['Predection_Manual']=50122.192990+-0.026816*Startups['Administration']-0.805715*Startups['Research']+0.027228*Startups['Marketing']
Startups['Predection_Manual']=model.fittedvalues
Startups['Error_Manual']=Startups['Profit']-Startups['Predection_Manual']
Startups['Errors_Model']=model.resid 
Startups
```

```python
print(model.tvalues, '\n',model.pvalues)
```

```python
(model.rsquared,model.rsquared_adj)
```
