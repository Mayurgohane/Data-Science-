```python
Bank= pd.read_csv("Bank.csv", sep=";")
```

```python
Bank.info()
```

```python

Bank.drop(["education"],inplace=True,axis = 1)
Bank.drop(["y"],inplace=True,axis = 1)
Bank.drop(["poutcome"],inplace=True,axis = 1)
```

```python
# Removing NA values in data set
Bank = Bank.dropna()
Bank.shape
```

```python
Bank
```

```python
# Dividing our data into input and output variables
X = Bank.iloc[:,1:]
Y = Bank.iloc[:,0]
```

```python
#Logistic regression and fit the model
classifier = LogisticRegression()
classifier.fit(X,Y)
```

```python
classifier.coef_
```

```python
Beta_Coefficients=pd.Series(classifier.coef_[0])
Beta_Coefficients
```

```python
Coef_df= pd.DataFrame({'Variables': ['balance','duration','campain','days','previous'],
                         'Beta Coefficients': Beta_Coefficients})
Coef_df
```

```python
classifier.intercept_
```

```python
classifier.classes_
```

```python
import numpy as np
np.set_printoptions(suppress=True)
classifier.predict_proba(X)[0:20]
```

```python
classifier.score(X,Y)
```

```python
#Predictions for X dataset
y_pred = classifier.predict(X)
```

```python
y_pred
```

```python
y_pred_df= pd.DataFrame({'actual': Y,
                         'predictions': classifier.predict(X)})
```

```python
y_pred_df[0:30]
```

```python

print(len(y_pred_df[y_pred_df['actual']==y_pred_df['predictions']]))
print(len(y_pred_df))

len(y_pred_df[y_pred_df['actual']==y_pred_df['predictions']])/len(y_pred_df)
```

```python
# Confusion Matrix for the model accuracy
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y,y_pred)
print(cm)
```

```python
TP=len(y_pred_df[(y_pred_df['actual']==1) & (y_pred_df['predictions']==1)])
TN=len(y_pred_df[(y_pred_df['actual']==0) & (y_pred_df['predictions']==0)])
FP=len(y_pred_df[(y_pred_df['actual']==0) & (y_pred_df['predictions']==1)])
FN=len(y_pred_df[(y_pred_df['actual']==1) & (y_pred_df['predictions']==0)])
```

```python
print('True Positives', TP)
print('True Negatives',TN)
print('False Positives', FP)
print('False Negatives', FN)
```
